```python
def is_prime():
  n = int(input("Enter a number to check for primality: "))

  if n <= 1:
    print("not a prime number")
  elif n == 2:
    print("prime number")
  else:
    is_a_prime = True
    for i in range(2, n):
      if n % i == 0:
        is_a_prime = False
        break
    if is_a_prime:
      print("prime number")
    else:
      print("not a prime number")

is_prime()
```

```python
def mergeSort(l,st,end):
    if st==end:
        return
    mid=(st+end)//2
    mergeSort(l,st,mid)
    mergeSort(l,mid+1,end)
    i=st
    j=mid+1
    l2=[]
    while i<=mid and j<=end:
        if l[i]<=l[j]:
            l2.append(l[i])
            i+=1
        else:
            l2.append(l[j])
            j+=1
    if i>mid:
        while j<=end:
            l2.append(l[j])
            j+=1
    elif j>end:
        while i<=mid:
            l2.append(l[i])
            i+=1
    for i in range(st,end+1):
        l[i]=l2.pop(0)
l=[2,8,1,4,9,8,2,6]
mergeSort(l,0,len(l)-1)
print(l)
```

```python
class Mobile:
    def __init__(self,songs,movies):
        self.volume=50
        self.brightness=40
        self.torch="Off"
        self.charge="Off"
        self.songs=songs
        self.movies=movies
    def playSongs(self):
        print(self.songs)

        inp=input("Please select one song from above list")
        if inp in self.songs:
            print(inp,"is playing")
        else:
            print("song doesnt exist and enter valid song")

    def playMovies(self):
        print(self.movies)

        inp=input("Please select one movie from above list")
        if inp in self.movies:
            print(inp,"is playing")
        else:
            print("movie doesnt exist and enter valid movie")

    def incVolume(self):
        inp=int(input("How much volume you want to increase"))
        if self.volume+inp>100:
            self.volume=100
        else:
            self.volume+=inp
        print("Volume increased successfully curr vol is",self.volume)

    def decVolume(self):
        inp=int(input("How much volume you want to decrease"))
        if self.volume-inp<0:
            self.volume=0
        else:
            self.volume-=inp
        print("Volume decreased successfully curr vol is",self.volume)
    def incBrightness(self):
        inp=int(input("Enter how much brightness to be increased:"))
        if self.brightness+inp>100:
            self.brightness=100
        else:
            self.brightness+=inp
        print("Brightness increased successfully present brightness is",self.brightness)
    def decBrightness(self):
        inp=int(input("Enter how much brightness to be decreased:"))
        if self.brightness-inp<0:
            self.brightness=0
        else:
            self.brightness-=inp
        print("Brightness decreased successfully present brightness is",self.brightness)
    def torchOn(self):
        self.torch="On"
        print("Torch is in on mode")
    def torchOff(self):
        self.torch="Off"
        print("Torch is in off mode")
    def chargeOn(self):
        self.charge="On"
        print("mobile is connected to charge")
    def chargeOff(self):
        self.charge="Off"
        print("mobile is disconnected to charge")

songs=input("Enter some songs list:\n").split()
movies=input("Enter some songs list:\n").split()

mi=Mobile(songs,movies)


while True:
    opt=int(input('''
    Please select options
    1.increase volume
    2.decrease volume
    3.increase brightness
    4.decrease brightness
    5.torch on
    6.torch off
    7.charge on
    8.charge off
    9.play songs
    10.play movies
    11.shutdown/poweroff
    '''))
    if opt==1:
        mi.incVolume()
    elif opt==2:
        mi.decVolume()
    elif opt==3:
        mi.incBrightness()
    elif opt==4:
        mi.decBrightness()
    elif opt==5:
        mi.torchOn()
    elif opt==6:
        mi.torchOff()
    elif opt==7:
        mi.chargeOn()
    elif opt==8:
        mi.chargeOff()
    elif opt==9:
        mi.playSongs()
    elif opt==10:
        mi.playMovies()
    elif opt==11:
        print("mobile became shutdown successfully!")
        break
```

```python
class BankAccount:
    def __init__(self, account_number, holder_name, balance):
        self.account_number = account_number
        self.holder_name = holder_name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance")

    def check_balance(self):
        print("Current Balance:", self.balance)


class SavingsAccount(BankAccount):
    def __init__(self, account_number, holder_name, balance, interest_rate):
        super().__init__(account_number, holder_name, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate / 100
        self.balance += interest
        print("Interest Added:", interest)


# Object creation
s = SavingsAccount(12345, "Chaithu", 10000, 5)

s.deposit(2000)
s.withdraw(1500)
s.check_balance()
s.add_interest()
s.check_balance()
```

```python
,from collections import Counter

def is_prime(n):
    if n < 2:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True

s = input("Enter string: ")

freq = Counter(s)

result = []

for ch, count in freq.items():
    if is_prime(count):
        result.append(ch)

if result:
    print(", ".join(result))
else:
    print("No Character Found")
```

```python
from typing import List

class Solution:
    def bs(self,nums,target,i,j):
        if i>j:
            return -1
        mid=(i+j)//2
        if nums[mid]==target:
            return mid
        elif target<nums[mid]:
            j=mid-1
            res=self.bs(nums,target,i,j)
            return res
        else:
            i=mid+1
            res=self.bs(nums,target,i,j)
            return res
    def search(self, nums: List[int], target: int) -> int:
        res=self.bs(nums,target,0,len(nums)-1)
        return res
```
