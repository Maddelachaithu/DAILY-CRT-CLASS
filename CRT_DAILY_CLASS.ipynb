```python
def is_prime():
  n = int(input("Enter a number to check for primality: "))

  if n <= 1:
    print("not a prime number")
  elif n == 2:
    print("prime number")
  else:
    is_a_prime = True
    for i in range(2, n):
      if n % i == 0:
        is_a_prime = False
        break
    if is_a_prime:
      print("prime number")
    else:
      print("not a prime number")

is_prime()
```

```python
def findMax(l,idx):
    if idx==len(l)-1:
        return l[idx]
    maxx=l[idx]
    res=findMax(l,idx+1)
    if res>maxx:
        return res
    else:
        return maxx
l=[2,1,5,8,10]
a=findMax(l,0)

print(a)
```

```python
def BS(l,i,j,key):

    mid=(i+j)//2
    if l[mid]==key:
        return mid
    elif l[mid]<key:
        j=mid-1
        res=BS(l,i,j,key)
    else:
        i=mid+1
        res=BS(l,i,j,key)
```

```python
def BS(l,i,j,key):
    if i>j:
        return -1
    mid=(i+j)//2
    if l[mid]==key:
        return mid
    elif l[mid]>key:
        j=mid-1
        res=BS(l,i,j,key)
        return res
    else:
        i=mid+1
        res=BS(l,i,j,key)
        return res
l=[3,4,5,7,9,13,16]
a=BS(l,0,len(l)-1,19)
print(a)
```
